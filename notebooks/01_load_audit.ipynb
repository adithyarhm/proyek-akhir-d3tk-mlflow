```python
import pandas as pd
import numpy as np

from pathlib import Path
```

```python
project_root = Path().resolve().parent


df1 = pd.read_csv(project_root / 'data' / 'raw' / 'node1.csv')
df2 = pd.read_csv(project_root / 'data' / 'raw' / 'node2.csv')
df3 = pd.read_csv(project_root / 'data' / 'raw' / 'node3.csv')
df4 = pd.read_csv(project_root / 'data' / 'raw' / 'node4.csv')
df5 = pd.read_csv(project_root / 'data' / 'raw' / 'node5.csv')
df6 = pd.read_csv(project_root / 'data' / 'raw' / 'node6.csv')

```

```python
for i in range(1,7):
    print(f"cek head dari node{i}")
    print(eval(f"df{i}").head())
```

```python
## insert data dari bu giva

# node1
elevation1 = 2101
latitude1 = -7.166870
longitude1 = 107.401387
df1['Elevation'] = elevation1
df1['Latitude'] = latitude1
df1['Longitude'] = longitude1

# node2
elevation2 = 2195
latitude2 = -7.167397
longitude2 = 107.401775
df2['Elevation'] = elevation2
df2['Latitude'] = latitude2
df2['Longitude'] = longitude2

# node3
elevation3 = 2196
latitude3 = -7.167415
longitude3 = 107.402914
df3['Elevation'] = elevation3
df3['Latitude'] = latitude3
df3['Longitude'] = longitude3

#node4
elevation4 = 2193
latitude4 = -7.166614
longitude4 = 107.403483
df4['Elevation'] = elevation4
df4['Latitude'] = latitude4
df4['Longitude'] = longitude4

#node5
elevation5 = 2194
latitude5 = -7.166418
longitude5 = 107.404100
df5['Elevation'] = elevation5
df5['Latitude'] = latitude5
df5['Longitude'] = longitude5

#node6
elevation6 = 2200
latitude6 = -7.166833
longitude6 = 107.40411
df6['Elevation'] = elevation6
df6['Latitude'] = latitude6
df6['Longitude'] = longitude6
```

```python
for i in range(1,7):
    print(f"cek head dari node{i}")
    print(eval(f"df{i}").info())
```

```python
for i in range(1, 7):
    df = eval(f"df{i}")

    # ubah penamaan kolom
    df.rename(columns={
        'Node': 'node',
        'Location': 'location',
        'Weather': 'weather',
        'DateTime': 'datetime',
        'H2S': 'h2s',
        'SO2': 'so2',
        'Humidity': 'hum',
        'Temperature': 'temp',
        'WindsSpeed': 'windspeed',
        'Elevation': 'elevation',
        'Latitude': 'latitude',
        'Longitude': 'longitude'
    }, inplace=True)

    # convert tipe data
    df['datetime'] = pd.to_datetime(df['datetime'])
    df['h2s'] = df['h2s'].astype('float64')
    df['so2'] = df['so2'].astype('float64')
    df['hum'] = df['hum'].astype('float64')
    df['elevation'] = df['elevation'].astype('float64')

    print(f"setelah perubahan:")
    df.info()
```

```python
list_df = [df1, df2, df3, df4, df5, df6]
df_combined = pd.concat(list_df, ignore_index=True)
print("setelah digabungkan:")
df_combined.tail()
```

```python
df_combined.to_csv(project_root / 'data' / 'raw' / 'node_combined.csv', index=False)
```
